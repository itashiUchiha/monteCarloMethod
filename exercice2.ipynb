```python
#definir ntre fontion integrale
def f(x):
    return 4*np.sqrt(1-x**2)
#definir notre axe des y
#y_coord=np.random.uniform(0,4.0)
#definir l'axe de x entre -1 et 1 parck df=[-1,1]
x=np.linspace(-1,1,10000)
#produire les
y=f(x)
plt.plot(x,y)
# proba_lepoint ce trouve au de l'itntegrale= aire_integrale/aire_rectangle

  
```

```python
def test():
    surface_rec=1*4
    count=0.0
    in_area=0.0
    listn=[10000,100000,1000000,10000000]
    surface_all=[]
    for i in listn:
        while count<i:
            x_coord=np.random.uniform(0,1.0)
            y_coord=np.random.uniform(0,4.0)
            if y_coord<f(x_coord):
                in_area+=1
            count+=1
        surface_integrale=(in_area/count)*(surface_rec)
        surface_all.append(surface_integrale)
    listb=[str(x) for x in listn]
    plt.plot(listb,surface_all)
    return surface_all
test()


```
